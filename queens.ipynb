```python
# is the queen safe?

def safety(matrix, row, col):
    
    # check columns
    
    for i in range(row):
        if matrix[i][col] == '1':
            return False
 
    # check lower left diagonal
    (i, j) = (row, col)
    while i >= 0 and j >= 0:
        if matrix[i][j] == '1':
            return False
        i = i - 1
        j = j - 1
 
    # check upper left diagonal
    (i, j) = (row, col)
    while i >= 0 and j < len(matrix):
        if matrix[i][j] == '1':
            return False
        i = i - 1
        j = j + 1
 
    return True
```

```python
#print the solution

def solution(matrix):
    for row in matrix:
        print(str(row).replace(',', '').replace('\'', ''))
    print()
```

```python
# function to iterate thru different placements

def queenfunction(matrix, row):
 
    # print if conditions are satisfied!
    if row == len(matrix):
        solution(matrix)
        return
 
    #place a queen in every spot and then check all other positions
    
    for i in range(len(matrix)):
 
        # check if queens are safe
        if safety(matrix, row, i):
            # if yes place the queen
            matrix[row][i] = '1'
 
            # do it again in the next row
            queenfunction(matrix, row + 1)
 
            # remove queen if it doesn't work
            matrix[row][i] = '–'
```

```python
# specify your N
N = 8

#make the matrix
matrix = [['–' for x in range(N)] for y in range(N)]
 
#apply the function starting on row 0

queenfunction(matrix, 0)
```
